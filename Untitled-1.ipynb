```python
import cv2
import os

# Create a directory to save frames
frames_dir = "D:/New folder/Assignment/object_frame"
os.makedirs(frames_dir, exist_ok=True)

# Load video
video_path = "/content/sample_data/video.mp4"  # Replace with your video file path
cap = cv2.VideoCapture(video_path)

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    frame_path = os.path.join(frames_dir, f"frame_{frame_count}.jpg")
    cv2.imwrite(frame_path, frame)

cap.release()
```

```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
import urllib.request

# Download Haar Cascade XML file
url = 'https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml?raw=true'
urllib.request.urlretrieve(url, 'haarcascade_frontalface_default.xml')

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Check if the Haar Cascade file was loaded correctly
if face_cascade.empty():
    raise ValueError("Haar Cascade XML file not loaded correctly.")

# Upload or download image file
# uploaded = files.upload()  # This will allow you to upload an image file

# Load the image
img = cv2.imread('/content/drive/MyDrive/objectsFrames/frame_1164.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert the image to grayscale

# Perform face detection
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw bounding boxes around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Draw a rectangle around the face

# Save or display the image with detected faces
cv2.imwrite('/content/drive/MyDrive/detected_objects/output.jpg', img)  # Save the result

# Display the result in Colab
cv2_imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Use cv2_imshow instead of imshow in Colab
```

```python
import cv2
import os
import numpy as np
import urllib.request
from matplotlib import pyplot as plt

# Download Haar Cascade XML file
url = 'https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml?raw=true'
urllib.request.urlretrieve(url, 'haarcascade_frontalface_default.xml')

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Check if the Haar Cascade file was loaded correctly
if face_cascade.empty():
    raise ValueError("Haar Cascade XML file not loaded correctly.")


frame_count = 0
# Load the image (ensure the image file is in the same directory or specify the full path)

import glob
folder_path = 'D:/New folder/Assignment/object_frame/*'

# List all files in the folder
files = glob.glob(folder_path)

for A in files:
    img = cv2.imread(A)  # Change to the path of your image file
    if img is None:
        raise ValueError("Image file not found or could not be loaded.")

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert the image to grayscale

    # Perform face detection
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    if faces:
    # Draw bounding boxes around detected faces
        for (x, y, w, h) in faces:
            cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Draw a rectangle around the face

        # Save or display the image with detected faces
        # cv2.imwrite('output.jpg', img)  # Save the result

        # frame_count = 0
        frame_count += 1
        frame_path = os.path.join("D:/New folder/Assignment/detected_frame", f"detct_frame_{frame_count}.jpg")
        cv2.imwrite(frame_path, img)

        # Display the result using matplotlib
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert image from BGR to RGB
        plt.imshow(img_rgb)
        plt.axis('off')  # Hide the axes
        plt.show()

    else :
        print("does not detech any object ")

    
    
```
